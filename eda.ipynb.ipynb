```python
# relation of two numeric plot

sns.lmplot(data=df, x='fare', y= df[df.fare < 300])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the dataframe based on a condition
filtered_df = df[df['fare'] < 300]

# Create the lmplot with the filtered dataframe
sns.lmplot(data=filtered_df, x='age', y='fare', hue='pclass')

# Add labels and title
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Conditional LM Plot')

# Display the plot
plt.show()
```

# how to add column and make a age group

```python
sns.histplot(data=df, x='age')
```

```python
df.columns
```

```python
df['age_group'] = pd.cut(df['age'], [0,10,18,30,50,80], labels= ['kids', 'child', 'youth', 'middle age', 'old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_group")['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```

```python
df['family_size'] = pd.cut(df['parch'], [0,1,2,3,4,5], labels= ['small', 'mid', 'big', 'v big', 'alpha'])
```

```python
df.groupby("family_size")['parch'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='parch', vars=['age', 'fare', 'pclass'])
```
